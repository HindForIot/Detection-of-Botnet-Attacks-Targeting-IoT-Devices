# Get 15 packets from each flow and store it in csv file

```python
b=['ip.len',
 'ip.flags',
 'ip.ttl',
 'ip.proto',
 'ip.checksum',
 'udp.srcport',
 'udp.dstport',
 'udp.length',
 'udp.checksum',
 'udp.time_relative',
 'udp.time_delta',
 'tcp.dstport',
 'tcp.srcport',
 'frame.number']

for n in range(16,20):
    for i in b:
        a=i+str(n)
        del dfr[a]
```
