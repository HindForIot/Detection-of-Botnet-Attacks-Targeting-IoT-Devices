# Load data from a Log.Labeled file, with missing values handled as specified.

```python
import numpy as np
import pandas as pd
# Assign the filename
fpath = r"C:\Users\hind.makarem\Desktop\Server files\Test_folders\labeled\Capture-35-1.labeled"
# Import file using np.recfromcsv: d
narray = np.recfromcsv(fpath, encoding=None)

# Print out first n records of d 
print(narray)
# string to search in file
word = 'fields'
with open(fpath, 'r') as fp:
    # read all lines in a list
    lines = fp.readlines()
    for line in lines:
        # check if string present on a current line
        if line.find(word) != -1:
            print(word, 'string exists in file')
            print('Line Number:', lines.index(line))
            print('Line:', line)
            linefields = line
x = linefields.split("\t")
df = pd.concat([pd.Series(l.astype(str).split('\t')) for l in narray], axis=1).T
print(df)
```
